# Movie Recommender System:

A Movie Recommender System is a machine learning project designed to suggest movies to users based on their preferences and behavior. The system utilizes various algorithms and techniques to analyze user data, such as movie ratings, genre preferences, and viewing history, to generate personalized recommendations.## 

The recommender system typically consists of three main components:

1. **User Interface:** Allows users to interact with the system, providing feedback such as ratings, likes, and dislikes.

2. **Recommendation Engine:** Uses machine learning algorithms to process user data and generate movie recommendations. Common algorithms include collaborative filtering, content-based filtering, and hybrid methods that combine both approaches.

3. **Database:** Stores information about movies, user preferences, and other relevant data needed for the recommendation engine to operate efftisfaction.

```python
df = df[['title', 'movie_id', 'cast', 'crew', 'original_language',
         'production_companies', 'overview',
         'popularity', 'genres', 'keywords']]
df.head()
```

```python
import ast
```

```python
df.head()
```

```python
def for_genere(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def for_keywords(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def for_production_companies(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
df.genres = df.genres.apply(for_genere)
df.keywords = df.keywords.apply(for_keywords)
df.production_companies = df.production_companies.apply(for_production_companies)
```

```python
df.head()
```

```python
df['crew'].iloc[0]
```

```python
def for_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
df.crew = df.crew.apply(for_director)
```

```python
df.head()
```

```python
def for_cast(obj):
    counter = 0
    L = []
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
df.cast = df.cast.apply(for_cast)
```

```python
df.head()
```

```python
df.overview = df.overview.apply(lambda x:x.split())
```

```python
df.head()
```

```python
change = lambda x : [i.replace(" ", "") for i in x]
```

```python
df.popularity = df.popularity.astype(str)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])[1:6]
```

```python
def recomend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse=True, key = lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
